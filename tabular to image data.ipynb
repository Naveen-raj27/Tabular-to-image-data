```python
import os
import pandas as pd
import numpy as np
from skimage import transform
import imageio
import matplotlib.pyplot as plt  # Import matplotlib for image display

# Load your CSV file
file_path = 'log2.csv'
df = pd.read_csv(file_path)  # Corrected variable name from df1 to df
print(df.iloc[1:11])  # Corrected indexing to iloc

# Mapping actions to numerical labels
action_mapping = {'allow': 0, 'deny': 1, 'drop': 2}  # Update with your actual labels
df['Action'] = df['Action'].map(action_mapping)

# Extract features and target
X = df.drop('Action', axis=1).values
y = df['Action'].values

# Normalize the data
X = (X - X.mean()) / X.std()

# Reshape the data into a 3D array
X_3d = X.reshape(-1, 1, X.shape[1])

# Function to apply IGTD transformation
def apply_igtd(data):
    return transform.integral_image(data)

# Apply IGTD transformation to the data
X_transformed = np.apply_along_axis(apply_igtd, axis=0, arr=X_3d)

# Create a folder to store the images
output_folder = 'C:/Users/Naveen/Desktop/Image/'
os.makedirs(output_folder, exist_ok=True)

# Save the images to the folder
for i in range(X_transformed.shape[0]):
    image_path = os.path.join(output_folder, f'image_{i}.png')
    img = X_transformed[i, 0].squeeze()
    
    # Reshape to 2D before saving
    img_2d = img.reshape(img.shape[0], -1)
  
    imageio.imwrite(image_path, img_2d)

# Display one of the saved images
plt.imshow(X_transformed[0, 0, :, :].squeeze(), cmap='gray')
plt.title(f'Class: {y[0]}')
plt.show()
print(img_2d)
```
